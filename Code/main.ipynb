```python
import numpy as np
import pandas as pd
import f_plots as fplt

from scipy.stats import pearsonr
from datetime import datetime

# Import data
df = pd.read_excel('../Data/Dataset_EUA23.xlsx')

# Define the target variable (EUA)
eua = df['EUA']

# Define the date variable
date = df['Date']

# Remove the date column from the data for correlations
df_ = df.drop(df.columns[[0]], axis=1)

# Brexit event
br_date = datetime(2020,2,1)
br_value = 23.21

# Change of Phase 3 to Phase 4
ph_date = datetime(2021,1,1)
ph_value =  33.69

# Plotting
fplt.plot_eua(date, eua, br_date, br_value, ph_date)
```

```python
# Subset the Phase 3 data
X_3 = df_.iloc[0:1782,:]

# Compute the correlations toward the target variable (Phase 3)
correlations_3 = X_3.corrwith(df_["EUA"])

# Sort the correlations
correlations_3 = correlations_3.sort_values(ascending=False)

# Remove the first element of the correlations (it is the target variable)
correlations_3 = correlations_3[1:]

# Subset the Phase 4 data
X_4 = df_.iloc[1782:,]

# Compute the correlations toward the target variable (Phase 4)
correlations_4 = X_4.corrwith(df_["EUA"])

# Sort the correlations
correlations_4 = correlations_4.sort_values(ascending=False)

# Remove the first element of the correlations (it is the target variable)
correlations_4 = correlations_4[1:]


# Plotting the Phase 3 correlations
fplt.plot_corr_3(correlations_3)
```

```python
# Subset the Phase 4 data
X_4 = df_.iloc[1782:,]

# Compute the correlations toward the target variable (Phase 4)
correlations_4 = X_4.corrwith(df_["EUA"])

# Sort the correlations
correlations_4 = correlations_4.sort_values(ascending=False)

# Remove the first element of the correlations (it is the target variable)
correlations_4 = correlations_4[1:]

# Plotting the Phase 4 correlations
fplt.plot_corr_4(correlations_4)
```
